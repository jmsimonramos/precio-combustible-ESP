```python
import pandas as pd
import pickle
```

```python
estaciones = pd.read_csv("../data/EESS.csv", sep=";")
estaciones.head()
```

```python
umbral_admision = 20
n_estaciones = 15

recuento_estaciones = estaciones.groupby("Rótulo").size().reset_index(name="Total")\
    .sort_values(by="Total", ascending=False)
recuento_estaciones = recuento_estaciones[recuento_estaciones["Total"] >= umbral_admision]
recuento_estaciones = recuento_estaciones.iloc[:n_estaciones]

print("Número de marcas:", recuento_estaciones.shape[0])
recuento_estaciones.head()
```

```python
marcas_gasolineras = recuento_estaciones.Rótulo.values.tolist()
print(marcas_gasolineras)
```

```python
eess_filtradas = estaciones[estaciones["Rótulo"].isin(marcas_gasolineras)]
print(eess_filtradas.shape[0])
eess_filtradas.head()
```

```python
diccionario_marcas = {row["IDEESS"]: row["Rótulo"] for _, row in estaciones.iterrows() if row["Rótulo"] in marcas_gasolineras}
```

```python
with open("../data/diccionarioMarcasEESS.bin", "wb") as binaryFile:
    pickle.dump(diccionario_marcas, binaryFile)
```
